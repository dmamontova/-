# III. Числовые переменные

> ### Их в наших данных не так много, однако есть те, на которые было бы интересно посмотреть, а теперь подробнее

- ### Year

#### Давайте посмотрим на то, на какие года приходилось наибольшее количество конфликтов

```python
plt.figure(figsize=(11,7))
plt.grid(True,linewidth=0.4)
plt.hist(df['Year'],alpha = 0.6,bins=20)
plt.title('Распределение количества конфликтов по годам')
plt.ylabel('Количество конфликтов')
plt.xlabel('Год')
```

#### Можем заметить, что количество конфликтов резко возросло, начиная с 2010-х годов, наибольшее их количество произошло в период с 2018 по 2022 год
