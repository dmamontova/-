```python
df1['0'] = df1['0'].map(lambda x: x[:-4])
```

```python
import numpy as np
```

```python
dick  = df1.set_index('Unnamed: 0').T.to_dict('list')
dick2 = {}
for el in dick:
    dick2[(dick[el][0], dick[el][-1])] = dick[el][1]
```

```python
a = list()
for i in range(df.shape[0]):
    try:
        a.append(dick2[(df.iloc[i]['COUNTRY'], df.iloc[i]['YEAR'])])
    except:
        a.append(np.nan)
```

```python
df['index'] = a
```
