```python
import pandas as pd


df = pd.read_csv('all_data_indexes.csv', encoding = ' UTF-8')
df1 = pd.read_csv('classification.csv', encoding = ' UTF-8')
```

```python
cnt = dict()

for index, row in df1.iterrows():
    cnt[row[0]] = dict()
    for i in range(1987, 2024):
        cnt[row[0]][i] = row[i-1986]

```

```python
import numpy as np
```

```python
a = list()
for index, row in df.iterrows():
    try:
        a.append(cnt[row[19]][row[4]])
    except:
        a.append(np.nan)
```
